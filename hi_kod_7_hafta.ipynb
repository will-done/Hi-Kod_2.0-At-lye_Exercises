```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
veri = pd.read_csv("olimpiyatlar_veriseti.csv")
veri.head()
```

```python
veri.info()
```

```python
unique_etkinlik = pd.unique(veri.etkinlik)
print("eşsiz etkinlik sayısı: {}".format(len(unique_etkinlik)))
unique_etkinlik[:10]
```

```python
# Cinsiyet sayısını bulma
gender_counts = veri['cinsiyet'].value_counts()

# Histogram grafiği oluşturma
plt.bar(gender_counts.index, gender_counts.values, color=['red', 'yellow'])
plt.title('Cinsiyet Dağılımı')
plt.xlabel('Cinsiyet')
plt.ylabel('Sayısı')
plt.xticks(rotation=0)  # X eksenindeki etiketleri döndürme
plt.show()
```

```python
# Farklı grupları bulma
etkinlik_counts = veri['etkinlik'].value_counts()
print(etkinlik_counts)
```

```python
#histogram grafiği için fonk yazma
def plotHistogram(degisken):
    plt.figure()
    plt.hist(veri[degisken], bins =85, color= "yellow")
    plt.xlabel(degisken)
    plt.ylabel("Frekans")
    plt.title("Veri sıklığı - {}".format(degisken))
    plt.show()
```

```python
#tüm sayısal değişkenler için hist çizme
sayısal_degisken = ["yas", "boy","kilo","yil"]
for i in sayısal_degisken:
    plotHistogram(i)
```

```python
import matplotlib.pyplot as plt
plt.scatter(kadın.boy, kadın.kilo, alpha = 0.4, label = "kadın" ,color ="red")
plt.scatter(erkek.boy, erkek.kilo, alpha = 0.4, label = "erkek", color ="blue")
# Grafik başlık ve etiketleri
plt.title('Boy ve Kilo Arasındaki İlişki')
plt.xlabel('Boy (cm)')
plt.ylabel('Kilo (kg)')

# Grafiği gösterme
plt.show()
```

```python
import pandas as pd
# Cinsiyet ve madalya türlerine göre gruplama yaparak sayıları bulma
cinsiyete_gore_madalya = veri.groupby(['cinsiyet', 'madalya']).size().unstack()
# Sonuçları yazdırma
print(cinsiyete_gore_madalya)
```

```python
# Şehirlere ve madalya türlerine göre gruplama yaparak sayıları bulma
sehre_gore_madalya = veri.groupby(['sehir', 'madalya']).size().unstack()

# Sonuçları yazdırma
print(sehre_gore_madalya)
```

```python
# Yıllara göre boy, kilo ve yaş ortalamalarını bulma
yil_ortalaması = veri.groupby('yil')[['boy', 'kilo', 'yas']].mean()

# Sonuçları yazdırma
print(yil_ortalaması)
```

```python
# Yıllara ve ülkelere göre madalya sayısını bulma
yillara_gore_spor_madalyaları = veri.groupby(['yil', 'spor', 'madalya']).size().unstack(fill_value=0)

# Sonuçları yazdırma
print(yillara_gore_spor_madalyaları)
```
